```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import SolveSudoku
import Img2Arr

```

```python
# Read the image
img = cv2.imread('sudoku1.jpg')
plt.imshow(img)
```

```python
# Img2Arr
arr = Img2Arr.img2arr(img)
```

```python
arr
```

```python
# SolveSudoku
solved = SolveSudoku.SolveSudoku(arr)
```

```python
solved
```

```python
solved_t = np.transpose(solved)
for i in range(9):
    print(solved_t[i])
```

```python
# Display this array as an image of grid of numbers like a sudoku
def display_sudoku(grid):
    fig, ax = plt.subplots()

    # Create a 9x9 grid
    for i in range(10):
        if i % 3 == 0:
            linewidth = 2
        else:
            linewidth = 1

        ax.plot([i, i], [0, 9], color='black', linewidth=linewidth)
        ax.plot([0, 9], [i, i], color='black', linewidth=linewidth)

    # Fill in the numbers
    for i in range(9):
        for j in range(9):
            if grid[i][j] != 0:
                ax.text(j + 0.5, 8.5 - i, grid[i][j], ha='center', va='center', fontsize=16)

    ax.axis('off')
    plt.show()

display_sudoku(solved_t)
```
