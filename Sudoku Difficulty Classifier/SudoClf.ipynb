```python
count_zeros = []
with open('sudokus.txt') as f:
    for line in f:
        count_zeros.append(line.split()[1].count('0'))
```

```python
import matplotlib.pyplot as plt

plt.hist(count_zeros, bins=range(0, 82, 1), edgecolor='black')
plt.xlabel('Number of zeros')
plt.ylabel('Frequency')
plt.title('Number of zeros in the sudokus')
plt.show()
```

```python
# Counting the number of sudokus with less than 50 zeros and classifying them as hard, the ones with 50-55 zeros as medium and the ones with more than 55 zeros as easy
hard = 0
medium = 0
easy = 0

for i in count_zeros:
    if i < 50:
        hard += 1
    elif i < 55:
        medium += 1
    else:
        easy += 1

print('Hard:', hard)
print('Medium:', medium)
print('Easy:', easy)
```

```python
# Reading the sudokus from the file and storing them in a list based on their difficulty
hard_sudokus = []
medium_sudokus = []
easy_sudokus = []

with open('sudokus.txt') as f:
    for line in f:
        sudoku = line.split()[1]
        zeros = line.split()[1].count('0')
        if zeros < 50:
            hard_sudokus.append(sudoku)
        elif zeros < 55:
            medium_sudokus.append(sudoku)
        else:
            easy_sudokus.append(sudoku)

# Writing the sudokus to files
with open('easy_sudokus.txt', 'w') as f:
    for sudoku in easy_sudokus:
        f.write(sudoku + '\n')

with open('medium_sudokus.txt', 'w') as f:
    for sudoku in medium_sudokus:
        f.write(sudoku + '\n')

with open('hard_sudokus.txt', 'w') as f:
    for sudoku in hard_sudokus:
        f.write(sudoku + '\n')

```
